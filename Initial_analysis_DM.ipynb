```python
############################################factor_1###########################################################################
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
from wordcloud import WordCloud, STOPWORDS

drive.mount('/content/drive')
link = 'https://drive.google.com/file/d/12cTH86zkyU8SFG6lh1vYx3kalDj5p59j/view?usp=share_link'

df = pd.read_csv('/content/drive/MyDrive/Crashes.csv' ,usecols=['date','time','factor_1'],parse_dates=["date"])

df_list=df['factor_1'].tolist()
df_list=list(filter(lambda factor: factor != ' ', df_list))
df_list=list(filter(lambda factor: factor != 'NO APPARENT CONTRIBUTING FACTOR', df_list))
df_list=list(filter(lambda factor: factor != 'NONE APPARENT', df_list))

df_filter=pd.DataFrame(df_list, columns=['factor_1'])
df = df_filter.replace('UNKNOWN',np.nan)
df.dropna(axis=0, how='any',inplace=True)

plt.figure(figsize=(10, 6), dpi=90)
df.factor_1.value_counts().plot(kind='bar')
plt.xlabel("Factors of accident")
plt.ylabel("No. of accidents")
```
